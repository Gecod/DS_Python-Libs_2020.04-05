#### 9). Создайте список high_corr из признаков, корреляция которых с полем target по абсолютному значению превышает 0.5 (причем, само поле target не должно входить в этот список).

```python
# список значений корреляции признаков с признаком target, кроме самого target
X_corr_to_target = X_corr.values[-1, :-1]
X_corr_to_target
```

```python
# индексы значений корреляции из предыдущего списка по абсолютному значению превышающих 0.5
ind_XtT = np.where(abs(X_corr_to_target) > 0.5)
print(ind_XtT)
```

```python
# создаем новый список из списка названий признаков по индексам из предыдущего списка
high_corr0 = list(X_corr.columns[ind_XtT])
high_corr0
```

```python
# или одним действием
high_corr = list(X_corr.columns[np.where(abs(X_corr.values[-1, :-1]) > 0.5)])
high_corr
```
